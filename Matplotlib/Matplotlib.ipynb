```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0, 5, 11)
y = x ** 2
```

```python
x
```

```python
y
```

```python
plt.plot(x, y)
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('Title')
```

```python
plt.subplot(1, 2, 1)
plt.plot(x, y, 'r')

plt.subplot(1, 2, 2)
plt.plot(y, x, 'b')
```

```python
fig = plt.figure()

axes = fig.add_axes([0.1, 0.1, 0.8, 0.8]) # (in percentage) left, bottom, width, height

axes.plot(x, y)
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('Set Title')
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes2 = fig.add_axes([0.5, 0.2, 0.4, 0.3])
axes3 = fig.add_axes([0.5, 0.6, 0.4, 0.3])
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes2 = fig.add_axes([0.2, 0.5, 0.4, 0.3])

axes1.plot(x, y)
axes1.set_title('Larger Plot')

axes2.plot(y, x)
axes2.set_title('Smaller Plot')
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2)

axes[0].plot(x,y)
axes[0].set_title('First plot')

axes[1].plot(y,x)
axes[1].set_title('Second plot')
```

```python
fig = plt.figure(figsize=(8,2)) # width and height

ax = fig.add_axes([0, 0, 1, 1])
ax.plot(x,y)
```
